```python
import matplotlib.pyplot as plt
```

```python
# Overs list
overs = list(range(1, 51))

# Newzeland runs 
nz = [4, 6, 10, 26, 37, 46, 51, 58, 64, 69, 73, 75, 77, 81, 83, 85, 90, 95, 99, 101, 102, 104, 107, 110, 114, 116, 123, 126, 131, 135, 138, 143, 147, 149, 153, 156, 161, 165, 168, 172, 175, 178, 184, 196, 201, 212, 216, 227, 239, 251]

# IND runs
ind = [9, 22, 25, 30, 31, 39, 45, 59, 60, 64, 65, 71, 76, 86, 93, 97, 100, 103, 106, 108, 113, 115, 117, 121, 122, 122, 122, 128, 133, 136, 141, 145, 154, 159, 161, 166, 176, 183, 190, 191, 202, 203, 206,212, 220, 231, 240, 245, 254]

# Wickets - Overs where each team lost wickets
nz_wickets = [8,11,13,24,38,46]  # nz wickets at these overs
ind_wickets = [19,20,27,39,42]  # IND wickets at these overs

# Ensure we only mark wickets where data exists
nz_wickets_valid = [o for o in nz_wickets if o <= len(nz)]
ind_wickets_valid = [o for o in ind_wickets if o <= len(ind)]
```

```python
# Create a figure
plt.figure(figsize=(12, 6))

# Plot runs for both teams
plt.plot(range(1, len(nz) + 1), nz, marker='o', linestyle='-', label='NZ', color='black')
plt.plot(range(1, len(ind) + 1), ind, marker='s', linestyle='-', label='IND', color='blue')

# Mark all wickets
plt.scatter(nz_wickets_valid, [nz[o-1] for o in nz_wickets_valid], color='red', marker='x', s=100, label=" Wickets")
plt.scatter(ind_wickets_valid, [ind[o-1] for o in ind_wickets_valid], color='red', marker='x', s=100,)

# Labels and title
plt.xlabel("Overs ")
plt.ylabel("Runs ")
plt.title("Cricket Score Comparison:Final NZ vs IND")

# Show legend
plt.legend()

# Show grid
plt.grid(True, linestyle="--", alpha=0.5)

# Display the plot
plt.show()
```
